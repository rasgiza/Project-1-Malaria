## Plot a bar graph to analyze which country recieves the most malaira net on average

```python
#set attribute to how the plot will display
fig=plt.figure(figsize=(16,10))
ax= fig.gca()#define axis
# numberline=np.linspace(0,5000000,10000)
ax.set_title('Bar plot of countries vs the number of malaria net distributed')# Give the plot the main title
# set text for the x axis
ax.set_ylabel('Number of net distributed')# set text for the y axis
ax.set_xlabel('Countries')
df1_group.plot.bar()
```
